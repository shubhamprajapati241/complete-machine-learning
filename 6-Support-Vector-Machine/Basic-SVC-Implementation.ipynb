## Support Vector Classified (SVC) Implementation

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pylab as plt 
```

```python
## Lets create sythetic datapoints
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0)
```

```python
X
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(X)[0] # taking only first column
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1], hue=y)
```

```python
## Data Splitting into train and test
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel="linear") # When datapoints are clearly seperable then use linear kernel

svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix 
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred)) 

# Becuase datapoints are easily seperable, that's why 99% accuracy
```

### With RBF, Polynomial, sigmoid kernel 
- Each and every kernel apply different mathematical formulas on dataset

```python
X,y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1], hue=y)
```

```python
## With Linear kernel
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=10)

svc = SVC(kernel="linear")

svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred)) 
```

```python
svc.coef_
```

```python
svc.intercept_
```

```python
## With rbf kernel
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=10)

svc = SVC(kernel="rbf") 

svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred)) 
```

```python
## With poly kernel
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=10)

svc = SVC(kernel="poly") # When datapoints are clearly seperable then use linear kernel

svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred)) 
```

```python
## With sigmoid kernel
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=10)

svc = SVC(kernel="sigmoid") # When datapoints are clearly seperable then use linear kernel

svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred)) 
```

Over here, RBF kernel is giving good accuracy as compared to all

#### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV 

# defining parameter range 
param_grid = {'C' : [0.1, 1, 10, 100, 1000],
              'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel' : ['rbf']
              }

grid = GridSearchCV(SVC(), param_grid=param_grid, cv=5, verbose=3, refit=True)
grid.fit(X_train, y_train)
```

```python
grid.best_params_ ## Best parameters
```

```python
y_pred = grid.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred)) 
```
